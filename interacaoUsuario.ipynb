Type

```python
x = 10
type(x)
print('x', type(x))

nome = 'Aryelle'
type(nome)
print('nome', type(nome))

num = '100'
num_convert = int(num)    # converter variável
print('num', type(num))
print('num_convert', type(num_convert))

num_float = float(num)
type(num_float)
print('num_float', type(num_float))
```
