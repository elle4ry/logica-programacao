# Condicional

```python
# If
if 5 < 10:
  print('Verdade')
```

```python
#   If/else
num1 = 10
num2 = 5
if num1 < num2:
  print('num 2 é maior')
else:
  print('num1 é maior')
```

```python
# if/ elif/ else
num1 = 10-8
num2 = 5+5
if num1 < num2:
  print('num2 é maior')
elif num1 == num2:
  print('Igual')
else:
  print('num1 é maior')
```

# Condicional com operadores lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if num1 > num2 and num1 > num3:
  print('num1 é maior')
else:
  print('num1 não é maior')
```
