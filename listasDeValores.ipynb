# Lista

```python
lista = [5, 'Aryelle', 8.9]
```

```python
type(lista)
```

```python
lista.append('Francisco') #Adiciona um elemento na lista
```

```python
print(lista)
print(lista[1])
```

# Tupla

```python
tupla = (5, 'Aryelle', 8.9)
```

```python
type(tupla)
```

```python
tupla.append('Francisco') #Não é possível adicionar elementos em uma tupla
```

```python
print(tupla)
```

# Dicionário

```python
dicionario = {'nome': 'Aryelle', 'cidade': 'muriae'}
```

```python
type(dicionario)
```

```python
print(dicionario)
```

```python
print(dicionario['nome'])
```

```python
dicionario['nome'] = 'Francisco'
```

```python
print(dicionario)
```
